```python
# PRACTICAL 1: Student Records Using Lists and Tuples
students = [
("Alice", [70, 80, 90]),
("Brian", [60, 75, 85]),
("Cynthia", [88, 92, 95])
]
for student in students:
 name, marks = student
 average = sum(marks) / len(marks)
 print("Student:", name, "| Average Score:", average)
```

```python
# PRACTICAL 2: Dictionary-Based Student Summary
student_scores = {
     "Alice": [70, 80, 90],
    "Brian": [60, 75, 85],
    "Cynthia": [88, 92, 95]
}

summary = {}

for name, scores in student_scores.items():
    summary[name] = {
        "total": sum(scores),
        "average": sum(scores) / len(scores)
    }

for name, data in summary.items():
    print(name)
    print("  Total:", data["total"])
    print("  Average:", data["average"])
```

Q1:Explain the difference between a list and a tuple

Lists are ordered and mutable, while tuples are ordered but immutable. Lists allow modification, tuples prevent accidental changes.

Q2:When would you use a dictionary instead of a list?

When data has named attributes or requires fast lookups. Example: storing student records with keys like “name” and “grade.”

Q3:Why are sets useful in data science?

They remove duplicates automatically and allow fast membership testing, which is crucial in cleaning datasets.


Q4: What happens if you try to modify a tuple

Python raises a TypeError because tuples are immutable.

```python
#Activity1
students = [
    ("Alice", [70, 80, 90]),
    ("Brian", [60, 75, 85]),
    ("Cynthia", [88, 92, 95]),
    ("David", [75, 85, 80])
]

for student in students:
    name, marks = student
    total = sum(marks)
    average = total / len(marks)
    print("Student:", name)
    print(" Total:", total)
    print(" Average:", average)
    print()
```

```python
#Activity2
student_scores = {
    "Alice": [70, 80, 90],
    "Brian": [60, 75, 85],
    "Cynthia": [88, 92, 95]
}

summary = {}

for name, scores in student_scores.items():
    summary[name] = {
        "total": sum(scores),
        "average": sum(scores) / len(scores),
        "max": max(scores)
    }

for name, data in summary.items():
    print(name)
    print(" Total:", data["total"])
    print(" Average:", data["average"])
    print(" Max:", data["max"])
    print()
```
